```python
import pandas as pd
from apyori import apriori

# Load and preprocess the Iris dataset
url = "iris.csv"
#columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris = pd.read_csv(url, header=0, names=columns)

# Discretize the features into categories (low, medium, high)
for col in columns[:-1]:  # Exclude 'class'
    iris[col] = pd.cut(iris[col], bins=3, labels=['low', 'medium', 'high'])

# Create the transaction list
transactions = iris.map(str).values.tolist()

# Apply Apriori algorithm
rules = apriori(transactions, min_support=0.2, min_confidence=0.6, min_lift=1.0, min_length=2)
result = list(rules)

# Display the rules
for rule in result:
    print(f"Rule: {', '.join(rule.items)}")
    print(f"Support: {rule.support}")
    for ordered_stat in rule.ordered_statistics:
        print(f"Confidence: {ordered_stat.confidence}, Lift: {ordered_stat.lift}")
    print("=" * 40)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
url="iris.csv"
columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris=pd.read_csv(url,header=0,names=columns)
iris['sepal_length']=pd.cut(iris['sepal_length'],bins=3,labels=['low','medium','high'])
iris['sepal_width']=pd.cut(iris['sepal_width'],bins=3,labels=['low','medium','high'])
iris['petal_length']=pd.cut(iris['petal_length'],bins=3,labels=['low','medium','high'])
iris['petal_width']=pd.cut(iris['petal_width'],bins=3,labels=['low','medium','high'])
transaction=[]
for i in range(0,len(iris)):
    transaction.append([str(iris.iloc[i,j]) for j in range(iris.shape[1])])
rules=apriori(transaction,min_support=0.2,min_confidence=0.6,min_lift=1.0,min_length=2)
result=list(rules)
for rule in result:
    print(f"Rule: {rule.items}")
    print(f"Support: {rule.support}")
    for ordered_stat in rule.ordered_statistics:
        print(f"Confidence: {ordered_stat.confidence}")
        print(f"Lift: {ordered_stat.lift}")
    print("="*40)
```
